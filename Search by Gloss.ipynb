```python
def searchgloss(base, glosses, **kwargs):
    query = 'search_sent?' + f'n_words=1&random_seed=&wf1=&lex1=&gr1=&gloss_index1={"-".join(glosses).upper()}&w_id1=&l_id1=&languages=ady&page_size=10&sort=random&distance_strict=on'
    return requests.get(base + query, **kwargs).text
```

```python
print(searchgloss(base, ['stem', 'obl'], cookies=cookie))
```
